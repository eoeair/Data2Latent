```python
import jax
import jax.numpy as jnp

data = jnp.load('data/mnist.npz')
x_train, y_train, x_test, y_test = data.values()
```

pca visual

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# 1. PCA2
pca = PCA(n_components=2)
x_train_embedded = pca.fit_transform(x_train.reshape(x_train.shape[0], -1))

#  PCA 
plt.figure(figsize=(8, 6))
scatter = plt.scatter(x_train_embedded[:, 0], x_train_embedded[:, 1], c=y_train, cmap='tab10', alpha=0.6)
plt.colorbar(scatter)  
# 
plt.title("PCA Visualization of Global Dataset")
plt.xlabel("PCA component 1")
plt.ylabel("PCA component 2")
plt.show()
```

tsne visual

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# t-SNE
x_train_embedded = TSNE(n_components=2, random_state=42).fit_transform(x_train.reshape(x_train.shape[0],-1))

#  t-SNE 
plt.figure(figsize=(8, 6))
scatter = plt.scatter(x_train_embedded[:, 0], x_train_embedded[:, 1], c=y_train, cmap='tab10', alpha=0.6)
plt.colorbar(scatter)
# 
plt.title("t-SNE Visualization of Global Dataset")
plt.xlabel("t-SNE component 1")
plt.ylabel("t-SNE component 2")
plt.show()
```
